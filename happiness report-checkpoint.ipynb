```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()
```

```python
data_2 = lab_enc.fit_transform(data['Country'])
pd.Series(data_2)
```

```python
data_3 = lab_enc.fit_transform(data['Region'])
pd.Series(data_3)
```

```python
data['Country'] = data_2
data['Region'] = data_3
data
```

```python
y = data['Happiness Score']
x = data.drop(columns = ['Happiness Score'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# split data into train and test, model will be built on train data and tested on test data
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y, test_size=0.25,random_state=10)
```

```python
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
data.tail(2)
```

```python
#since we have already fit the scaler, we can transform the data
print('chance of admidssion:-',regression.predict(scaler.transform([[21,8,157,0.08658,0.01530,0.41587,0.22396,0.1185,0.10062,0.19727,1.83302]])))
```

```python
regression.score(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```

```python
#lets visualize the label predicted vs actual
y_pred = regression.predict(x_test)

plt.scatter(y_test,y_pred)
plt.xlabel('actual label')
plt.ylabel('predicted label')
plt.title('Actual vs predicted label')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
y_pred = regression.predict(x_test)
mean_absolute_error(y_test,y_pred)
#mean_squared_error(y_test,y_pred)
#np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
mean_squared_error(y_test,y_pred)
```

```python
#checking overfitting
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
```

```python
#lassovc will return best alpha values after max itteration
#normalize is subtracting the mean and dividing with l2 norm
lasscv = LassoCV(alphas=np.arange(0.001,0.1,0.01),normalize= True )# or lasscv = LassoCV(alphas =None,max_iter= (#anynumber u want),normalize = Trrue)
lasscv.fit(x_train,y_train)
```

```python
#best alpha parametres
alpha = lasscv.alpha_
alpha
```

```python
lasso_regg = Lasso(alpha)
lasso_regg.fit(x_train,y_train)
```

```python
lasso_regg.score(x_test,y_test)
```
